```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df=pd.read_csv('/content/drive/My Drive/2011-01.csv')
```

```
df.head(5)
```

```
Start_station_n = pd.DataFrame(df['Start station number'])
End_station_n = pd.DataFrame(df['End station number'])
```

```
plt.figure()
plt.scatter(Start_station_n,End_station_n)
plt.xlabel('Start station number')
plt.ylabel('End station number')
plt.show()
```

```
feature_cols = ['Start station number','End station number']
x = df[feature_cols]
y = df['End station number']
lm = LinearRegression()
lm.fit(x,y)
```

```
print(lm.intercept_)
print(lm.coef_)
```

```
X = df['Start station number'].values.reshape(-1,1)
y = df['End station number'].values.reshape(-1,1)
```

```
from sklearn.model_selection import train_test_split 
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```
y_pred = lm.predict(X_test)
```
